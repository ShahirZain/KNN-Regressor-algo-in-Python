```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split  
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error 
from math import sqrt
mainTable = pd.read_csv("periodic_table.csv",encoding='iso-8859-1', skiprows=[92,99,70,56,38,2])
SubTable = mainTable[['number','atomic_number','period']]
SubTable
```

```python
X = SubTable.iloc[:, SubTable.columns != 'atomic_number'].values  
Y = SubTable.iloc[:,SubTable.columns == 'atomic_number']
X_train, X_test, y_train, y_test = train_test_split(X, Y, shuffle=True ,random_state=42) 
```

```python
len(X_train),len(X_test)
```

```python
rmse_val = []
for k in range(20):
    k = k + 1
    neigh = KNeighborsRegressor(n_neighbors=k )
    neigh.fit(X_train, y_train)
    predicted = neigh.predict(X_test)
    predicted = np.round(predicted)
    error = sqrt(mean_squared_error(y_test,predicted)) 
    rmse_val.append(error) 
    print('RMSE value for k= ' , k , 'is:', error)
```

```python
curve = pd.DataFrame(rmse_val)
curve.plot()
```
